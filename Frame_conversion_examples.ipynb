# Frame conversion example code
The intention of these plots are to help make clear the following:
1. Show how the helio and geo projected frames are related by the parallax vector at t0par.
2. Visualize the differences between helio and geo projected frames.
3. Visualize the difference between source-lens and lens-source proper motion.
4. Visualize the difference between East-North and tau-beta coordinate systems.
5. Show how the conventions of 2., 3., and 4. are all independent of each other.

From top to bottom:
1. Plot showing the conversion between heliocentric and geocentric projected parameters, using the Lu convention (i.e. the signs of piEE and piEN are for source-lens relative proper motion, and the sign of u0 is using the East-North coordinate system). The arrows pointing from the source (origin) to the lens are the u0 vector. The arrow perpendicular to that is the direction of the source-lens relative proper motion. The length of the arrow is (t0par - t0)/tE. In this case, since t0par = t0_helio, the length of the arrow is zero. The gray arrow is the parallax vector at t0par. The sign of u0 is determined as follows: if the East component of the u0 vector is positive, u0 is positive; if the East compoenent of the u0 vector is negative, then u0 is negative. Here you can see that the sign of u0 can differ between the two frames, as well as the piE vector.
2. Same as 1., but in the Gould convention (i.e. the signs of piEE and piEN and for the lens-source relative proper motion, and the sign of u0 is using the tau-beta coordinate system). The arrows pointing from the lens (origin) to the source are the u0 vector. The arrow perpendicular to that is the direction of the lens-source relative proper motion. The length of the arrow is (t0par - t0)/tE. The small plot to the right shows the tau-beta coordinate system. It is always right-handed with respect to (tau, beta). The sign of u0 is determined as follows: if beta and u0 vectors are in the same direction, u0 is positive; if beta and u0 vectors are in opposite directions, u0 is negative. Here you can see that the sign of u0 again differs between the two frames, as well as the piE vector. 
3. Plot showing the lightcurves of the models calculated by BAGLE and MulensModel, using the values in the printout (which correspond to the plots above). This is to validate that the conversion is correct.
